```python
import pandas as pd
import numpy as np

# Leer datos Clase/codes/datos_5_ej5.csv
df_datos = pd.read_csv('../Clase/codes/datos_5_ej5.csv') 
df_datos.head()
```

```python
# Convertir y a numpy array
y = df_datos['y'].values

# Convertir x1 a x5 a numpy array en una matriz X y agregar una columna de unos para el término independiente
X = df_datos[['x1', 'x2', 'x3', 'x4', 'x5']].values
X = np.hstack((np.ones((X.shape[0], 1)), X))    
X
```

```python
# Obtener coeficientes de regresión lineal manualmente

# Calcular (X^T X)^(-1)
XtX_inv = np.linalg.inv(X.T @ X)
# Calcular (X^T y)
Xty = X.T @ y
# Calcular coeficientes: beta = (X^T X)^(-1) X^T y
beta = XtX_inv @ Xty
beta
```

```python
# Calcular Standard Error de los coeficientes
# Calcular predicciones
y_pred = X @ beta
# Calcular residuos
residuos = y - y_pred
residuos

# Calcular varianza de los residuos
var_residuos = np.sum(residuos**2) / (X.shape[0] - X.shape[1])
var_residuos

# Calcular varianza de los coeficientes: Var(beta) = var_residuos * (X^T X)^(-1)
var_beta = var_residuos * XtX_inv
var_beta

# Calcular Standard Error: SE(beta) = sqrt(Var(beta))
se_beta = np.sqrt(np.diag(var_beta))
se_beta
```

```python
# Calcular t-statistic: t = beta / SE(beta)
t_statistic = beta / se_beta
t_statistic
```
